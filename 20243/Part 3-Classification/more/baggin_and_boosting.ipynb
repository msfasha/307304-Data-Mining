# Random Forests, Bagging, and Boosting

## Random Forests

### Introduction

Random Forest is a powerful ensemble learning method that builds multiple decision trees and combines their results to improve predictive performance and control overfitting. It operates by constructing many decision trees during training time and outputting the class that is the mode of the classes (for classification) or the mean prediction (for regression) of the individual trees.

Each tree is trained on a different random subset of the data and considers a random subset of features when splitting nodes. This randomness ensures that the trees are diverse, making the Random Forest robust and highly accurate compared to individual decision trees.

### Benefits of Random Forests

- **Higher accuracy** than a single decision tree through aggregation.
- **Reduces overfitting** by averaging multiple trees, lowering variance.
- **Handles missing values** and remains accurate even when parts of the data are missing.
- **Works well with high-dimensional data** and large datasets.
- **Evaluates feature importance**, allowing insights into which variables most affect the predictions.

### Key Settings for Random Forests

- **n_estimators**: Number of decision trees in the forest.
- **Criterion**: Metric used to evaluate splits (e.g., 'gini' for Gini Impurity, 'entropy' for Information Gain).
- **Max Features**: Number of features to consider when searching for the best split.
- **Max Depth**: Maximum depth of the individual trees.
- **Bootstrap**: Whether to use bootstrapped samples to build trees.
- **Random State**: Controls randomness to ensure reproducibility.
- **OOB Score**: Whether to use out-of-bag samples to estimate model performance.

---

## Bootstrap Sampling in Random Forests

Bootstrap sampling is a central technique used in Random Forests. In bootstrap sampling:

- Each tree is trained on a **random sample of the dataset drawn with replacement**.
- Some instances may appear multiple times in a sample, while others may not appear at all.
- This procedure introduces diversity among the trees, helping to reduce overfitting.

If bootstrap sampling is disabled (`bootstrap=False`), each tree is trained on a **random subset without replacement**, a technique called **pasting**.

Thus, **bootstrap** and **pasting** are two ways to create diversity among the trees:
- **Bootstrap (default)**: Sampling **with replacement**.
- **Pasting**: Sampling **without replacement**.

The choice between bootstrap and pasting can affect model behavior:
- Bootstrap tends to produce more variance among trees, improving the ensemble's stability.
- Pasting uses different samples without duplication and may sometimes improve accuracy on specific datasets.

---

## Bagging (Bootstrap Aggregating)

### Introduction

Bagging, short for **Bootstrap Aggregating**, is an ensemble method that improves the stability and accuracy of machine learning algorithms. Bagging works by training multiple models in **parallel** on different subsets of the data and **aggregating their predictions**.

In bagging:
- Each model is trained independently on a **bootstrapped** subset (sampling with replacement).
- Final predictions are made by **voting** (classification) or **averaging** (regression).

### Benefits of Bagging

- **Reduces variance**: Independent models average out random fluctuations.
- **Helps prevent overfitting**: Especially effective with high-variance models like decision trees.
- **Improves stability**: Reduces sensitivity to noise in the training data.

### Configurable Settings in Bagging

- **n_estimators**: Number of models in the ensemble.
- **Bootstrap**: Whether to use bootstrapping (`True`) or pasting (`False`).
- **Max Samples**: Fraction or number of samples to draw for each model.
- **Max Features**: Number of features to draw for each model.
- **OOB Score**: Evaluate performance on samples not included in the bootstrap set.

**Random Forest** is a special case of bagging, where the base models are decision trees and additional randomness is introduced at feature splits.

---

## Boosting

### Introduction

Boosting is another ensemble method, but unlike bagging, it builds models **sequentially**. Each new model tries to correct the errors made by the previous models. Boosting focuses on **reducing bias** and achieving higher predictive accuracy.

In boosting:
- Each model is trained on the **entire dataset**, but places **more focus** on instances that previous models classified incorrectly.
- The predictions from all models are combined, typically through a **weighted sum**.

### Benefits of Boosting

- **Reduces bias**: Sequential learning improves model accuracy on difficult examples.
- **Produces strong learners**: From many weak learners.
- **Effective in a wide range of problems**: Especially those with complex decision boundaries.

### Configurable Settings in Boosting

- **n_estimators**: Number of boosting rounds.
- **Learning Rate**: Shrinks the contribution of each model, controlling how much each model influences the final prediction.
- **Max Depth**: Maximum depth of each weak learner (typically kept shallow).
- **Subsample**: Fraction of the training set used to train each model (can introduce randomness like in bagging).

Examples of boosting algorithms:
- **AdaBoost**: Focuses on misclassified examples by adjusting sample weights.
- **Gradient Boosting**: Models residual errors sequentially.
- **XGBoost, LightGBM, CatBoost**: Highly optimized gradient boosting libraries.

---

## Comparison Between Bagging and Boosting

| Feature               | Bagging                     | Boosting                      |
|------------------------|------------------------------|-------------------------------|
| Data Sampling          | Random subsets (bootstrapped) | Full dataset with weighting   |
| Model Training         | In parallel                  | Sequential                    |
| Main Focus             | Reduce variance              | Reduce bias                   |
| Use of Bootstrap       | Yes (in bagging)              | No (sampling usually without replacement or weighting) |
| Sensitivity to Noise   | Low                           | High (boosting can overfit if not regularized) |
| Example Algorithms     | Random Forest, BaggingClassifier | AdaBoost, Gradient Boosting |

# Random Forest Implementation using the Loan Default Dataset

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Load the dataset
url = "https://raw.githubusercontent.com/msfasha/307304-Data-Mining/main/datasets/loan_default.csv"
data = pd.read_csv(url)

# Separate features and target variable
X = data.drop('loan_default', axis=1)  # Features
y = data['loan_default']               # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Random Forest Classifier
rf_model = RandomForestClassifier(
    n_estimators=100,
    criterion='gini',
    max_depth=None,
    random_state=42,
    bootstrap=True  # Set to False to use pasting instead of bootstrapping
)

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Plot a single tree from the forest for visualization
plt.figure(figsize=(20, 10))
plot_tree(rf_model.estimators_[0], feature_names=X.columns, class_names=['No Default', 'Default'], filled=True)
plt.title("Random Forest - Single Decision Tree Visualization")
plt.savefig("random_forest_single_tree_plot.png", format="png", dpi=300, bbox_inches='tight')
plt.show()
```
