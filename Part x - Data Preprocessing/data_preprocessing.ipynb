## 1. Introduction to Data Preprocessing

**Brief Introduction:**  
Data preprocessing is a crucial step in any machine learning or data analysis project. It involves transforming raw, messy data into a clean and standardized format so that subsequent modeling efforts are both accurate and efficient. In this lecture, we will explore various preprocessing tasks such as data cleaning, feature transformation, and data splitting.

### Key Objectives
1. Understand why data preprocessing is important.
2. Learn methods to handle missing or erroneous data.
3. Apply encoding, scaling, and transformation on features.
4. Prepare data for modeling through splitting.

## 2. Overview of the Titanic Dataset

**Brief Introduction:**  
The Titanic dataset is one of the most popular datasets for learning data preprocessing and predictive modeling. It contains information about passengers aboard the RMS Titanic and whether they survived or not. This dataset is ideal because it includes a mix of numerical, categorical, and missing data, providing a realistic scenario for demonstrating preprocessing techniques.

### 2.1 Dataset Description
- **PassengerId**: A unique identifier for each passenger.
- **Survived**: Survival status (0 = No, 1 = Yes).
- **Pclass**: Passenger class (1 = First, 2 = Second, 3 = Third).
- **Name**: Name of the passenger.
- **Sex**: Gender of the passenger.
- **Age**: Age of the passenger.
- **SibSp**: Number of siblings/spouses aboard.
- **Parch**: Number of parents/children aboard.
- **Ticket**: Ticket number.
- **Fare**: Ticket fare.
- **Cabin**: Cabin number.
- **Embarked**: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).

### 2.2 Downloading the Dataset
The Titanic dataset is freely available on [Kaggle](https://www.kaggle.com/competitions/titanic). After downloading the CSV file (often named `train.csv`), you can store it in your working directory to follow along with the examples.

## 3. Initial Setup and Exploratory Analysis

**Brief Introduction:**  
Before we dive into data cleaning and transformation, it’s good practice to load the dataset, verify its contents, and perform some high-level checks. This helps us understand the data’s shape, the data types of each column, and identify potential issues (e.g., large numbers of missing values).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("train.csv")

# Inspect first few rows
print(df.head())

# View summary of the dataframe
print(df.info())

# Statistical summary for numerical features
print(df.describe())
```

### Observations to Make
1. **Number of rows vs. columns**: This tells us how big the dataset is.
2. **Column types**: Are they numerical, categorical, or mixed?
3. **Potential missing values**: Notice which columns have `NaN` or blank values.
4. **Data ranges**: Check for outliers or strange values using summary statistics.

---

## 4. Data Cleaning

**Brief Introduction:**  
Data cleaning often involves handling missing values, removing duplicates, and addressing outliers. The goal is to ensure that the dataset is as accurate and consistent as possible.

### 4.1 Handling Missing Values

1. **Identify Missing Data**  
   ```python
   df.isnull().sum()
   ```
   You might see that the `Age` and `Cabin` columns have missing values, and sometimes `Embarked` might also have a few missing entries.

2. **Strategies for Missing Data**  
   - **Drop rows** that have missing values.
   - **Impute (fill) missing values** using statistical methods (mean, median, mode) or advanced methods (predictive modeling, interpolation).

   ```python
   # Example of dropping rows where Cabin is missing (just for demonstration; often you'd keep it for feature engineering)
   df.dropna(subset=['Cabin'], inplace=True)

   # Imputing Age with the mean
   df['Age'].fillna(df['Age'].mean(), inplace=True)

   # Imputing Embarked with the mode (most frequent port)
   df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
   ```

### 4.2 Handling Duplicates

While duplicates are less likely in this particular dataset, it’s good to illustrate the process:

```python
# Check for duplicates
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")

# Drop duplicates if any
df.drop_duplicates(inplace=True)
```

### 4.3 Handling Outliers

Outliers can be identified using statistical metrics like the **Interquartile Range (IQR)** or via visual methods (e.g., box plots).

```python
# Visualize distribution of Fare
sns.boxplot(x=df['Fare'])
plt.show()

# Example: Remove outliers using IQR (demonstrative approach)
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df = df[(df['Fare'] >= lower_bound) & (df['Fare'] <= upper_bound)]
```

---

## 5. Data Transformation

**Brief Introduction:**  
Once data is cleaned, we often need to transform features. This can include encoding categorical variables, scaling numerical features, and creating new features that may be more predictive or meaningful.

### 5.1 Encoding Categorical Variables

Categorical features like `Sex` and `Embarked` may need to be represented numerically for certain models.

```python
# One-hot encoding
df = pd.get_dummies(df, columns=['Embarked'], prefix='Embarked')

# Label encoding for 'Sex' (example approach)
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])

# Check the transformed dataset
df.head()
```

### 5.2 Feature Scaling

Some models (e.g., distance-based models, neural networks) are sensitive to feature scale. Common scaling methods are **Standardization** (Z-score) and **Normalization** (Min-Max).

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()  # or MinMaxScaler()
df['Age_scaled'] = scaler.fit_transform(df[['Age']])
df['Fare_scaled'] = scaler.fit_transform(df[['Fare']])
```

### 5.3 Feature Creation (Optional but Useful)

You can create new features based on existing data. For instance, extracting a title from the `Name` column in the Titanic dataset might yield new insights.

```python
def extract_title(name):
    return name.split(',')[1].split('.')[0].strip()

df['Title'] = df['Name'].apply(extract_title)
df['Title'].value_counts()
```

## 6. Splitting the Data for Modeling

**Brief Introduction:**  
After cleaning and transforming the dataset, the final step in preprocessing is to split the data into training and testing sets. This ensures that we can train our model on one subset of the data and accurately evaluate its performance on unseen data.

```python
from sklearn.model_selection import train_test_split

# Features (X) and target (y)
X = df.drop(['Survived', 'Name', 'Ticket', 'Cabin', 'PassengerId'], axis=1)
y = df['Survived']

# Split data (70% training, 30% testing, for example)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=42)

# Verify the shape of the splits
print(X_train.shape, X_test.shape)
```
